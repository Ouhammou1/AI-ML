```python
import numpy as np
from scipy.optimize import minimize

observations = [1, 2, 0, 1, 2, 3, 1, 2, 0, 3]

def pmf(k, theta):
    if k == 0:
        return 2 * theta / 3
    elif k == 1:
        return theta / 3
    elif k == 2:
        return 2 * (1 - theta) / 3
    elif k == 3:
        return (1 - theta) / 3

def neg_log_likelihood(theta):
    val = -np.sum([np.log(pmf(k, theta)) for k in observations])
    return val

result = minimize(neg_log_likelihood, x0=0.5, bounds=[(0, 1)])
theta_mle = result.x[0]
print(f"Maximum Likelihood {theta_mle:.4f}")
```

```python
X = np.array([1, 5, 10, 15, 20, 22, 25, 28, 32, 34])
Y = np.array([1.9, 1.8, 1.7, 1.6, 1.6, 1.5, 1.4, 1.3, 1.3, 1.2])

n = len(X)
a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)
b = (np.sum(Y) - a * np.sum(X)) / n

print(f"Slope (a): {a:.4f}")
print(f"Intercept (b): {b:.4f}")
print(f"Equation of the line: Y = {a:.4f}X + {b:.4f}")
```
