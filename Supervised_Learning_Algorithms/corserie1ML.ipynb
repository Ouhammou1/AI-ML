Solution de l'Exercice 2 : Implémentation Python

Voici le code Python correspondant :

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Données
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5, 7])

# Modèle de régression linéaire
model = LinearRegression()
model.fit(X, y)

# Coefficients
beta_0 = model.intercept_  # Ordonnée à l'origine
beta_1 = model.coef_[0]    # Pente
print(f"Équation : y = {beta_0:.2f} + {beta_1:.2f}x")

# Droite de régression
y_pred = model.predict(X)

# Graphique
plt.scatter(X, y, color='blue', label='Données')
plt.plot(X, y_pred, color='red', label='Régression')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Prédiction pour x = 7
x_new = np.array([[7]])
y_new = model.predict(x_new)
print(f"Prédiction pour x=7 : y = {y_new[0]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# G´en´erer 50 points x al´eatoires dans l’intervalle [0, 10]
x = np.random.uniform(0, 10, 50)
# G´en´erer l’erreur epsilon selon une loi normale N(0, 1)
epsilon = np.random.normal(0, 1, 50)
# Calculer les valeurs de y
y = 2 * x + 1 + epsilon
plt.scatter(x, y, color='blue', label='Données')
#plt.plot(x, y, color='red', label='Donnees bruitees')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('exercice 4')
plt.show()

```

$\textbf{Ajustez une régression linéaire à ces données et tracez le graphique.}$

Nous utilisons la régression linéaire pour ajuster une droite aux données générées. Le modèle linéaire aura la forme : $ y = \theta_{0} + \theta_{1} x $.

Le processus d'ajustement consiste à trouver les coefficients $ \theta_0 $ et $\theta_1$ qui minimisent la somme des carrés des erreurs entre les valeurs prédites et les valeurs observées.

Voici le code Python pour ajuster la régression linéaire et tracer le graphique :

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Générer des données x et y avec bruit
np.random.seed(0)
x = np.random.uniform(0, 10, 50)
y = 2 * x + 1 + np.random.normal(0, 1, 50)  # y = 2x + 1 + bruit

# Reshaper les données x pour les adapter à la régression linéaire
x_reshaped = x.reshape(-1, 1)
print("x_reshaped",x_reshaped)
# Créer et ajuster le modèle de régression linéaire
model = LinearRegression()
model.fit(x_reshaped, y)

# Prédire les valeurs de y
y_pred = model.predict(x_reshaped)

# Tracer les données et la droite de régression
plt.scatter(x, y, color='blue', label='Données')
plt.plot(x, y_pred, color='red', label='Régression linéaire')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Régression Linéaire avec Bruit')
plt.show()
```

4. Analysez les r´esidus (diff´erence entre les y observ´es et les y pr´evus).

Les r´esidus repr´esentent l’´ecart entre les valeurs observ´ees y et les valeurs pr´edites
y pr´edit. On peut calculer les r´esidus comme suit : $y_{i} - y_i^{\wedge}$

residui = yi − yˆi
.
Les r´esidus doivent ˆetre analys´es pour v´erifier si le modele de regression lineaire est
bien ajust´e. Si les r´esidus sont al´eatoires et ne suivent pas de structure particuliere,
cela signifie que le modele est bien adapt´e aux donnees.

Voici le code pour calculer et afficher les r´esidus :


```python
# Calculer les résidus
residuals = y - y_pred

# Tracer les résidus
plt.scatter(x, residuals, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('Résidus')
plt.title('Analyse des Résidus')
plt.show()
```

Interprétation :
Si les résidus semblent aléatoires et ne présentent pas de tendance particulière, le modèle est bien ajusté. Sinon, des transformations ou des modèles plus complexes peuvent être nécessaires.

```python
import numpy as np

# Données
X = np.array([[50, 2], [60, 2], [70, 3], [80, 3], [90, 4]])  # Variables explicatives
y = np.array([200, 220, 250, 280, 310])  # Variable dépendante

# Ajouter une colonne de 1 pour le biais (intercept)
X_with_bias = np.c_[np.ones(X.shape[0]), X]

# Calcul des coefficients avec la formule des moindres carrés
# beta = (X'X)^(-1) X'y
beta = np.linalg.inv(X_with_bias.T @ X_with_bias) @ (X_with_bias.T @ y)

# Extraction des coefficients
beta_0, beta_1, beta_2 = beta

# Affichage des résultats
print(f"β₀ (intercept) = {beta_0}")
print(f"β₁ (coefficient pour la surface) = {beta_1}")
print(f"β₂ (coefficient pour le nombre de chambres) = {beta_2}")

# Prédiction pour une maison de 85 m² et 3 chambres
x_new = np.array([[85, 3]])
y_pred = model.predict(x_new)

print(f"Prédiction pour une maison de 85 m² et 3 chambres : {y_pred[0]:.2f} k€")
```

sol ex7 regression et biais variance

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fonction pour générer les données avec du bruit
def generate_data(n_points, noise_std):
    # Génération des valeurs de x
    x = np.linspace(0, 10, n_points).reshape(-1, 1)
    # Génération des valeurs de y selon la relation y = 3x + 2 avec un bruit ajouté
    noise = np.random.normal(0, noise_std, size=x.shape)
    y = 3 * x + 2 + noise
    return x, y

# Générer les données avec différents niveaux de bruit
x_low, y_low = generate_data(100, 0.5)  # Bruit faible
x_medium, y_medium = generate_data(100, 2)  # Bruit moyen
x_high, y_high = generate_data(100, 5)  # Bruit élevé

# Ajuster un modèle linéaire pour chaque jeu de données
model_low = LinearRegression().fit(x_low, y_low)
model_medium = LinearRegression().fit(x_medium, y_medium)
model_high = LinearRegression().fit(x_high, y_high)

# Prédictions sur les points x
y_pred_low = model_low.predict(x_low)
y_pred_medium = model_medium.predict(x_medium)
y_pred_high = model_high.predict(x_high)

# Tracer les résultats
plt.figure(figsize=(15, 5))

# Bruit faible
plt.subplot(1, 3, 1)
plt.scatter(x_low, y_low, color='blue', label='Données avec bruit faible')
plt.plot(x_low, y_pred_low, color='red', label='Modèle ajusté')
plt.title("Bruit Faible")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Bruit moyen
plt.subplot(1, 3, 2)
plt.scatter(x_medium, y_medium, color='blue', label='Données avec bruit moyen')
plt.plot(x_medium, y_pred_medium, color='red', label='Modèle ajusté')
plt.title("Bruit Moyen")
plt.xlabel('x')
plt.legend()

# Bruit élevé
plt.subplot(1, 3, 3)
plt.scatter(x_high, y_high, color='blue', label='Données avec bruit élevé')
plt.plot(x_high, y_pred_high, color='red', label='Modèle ajusté')
plt.title("Bruit Élevé")
plt.xlabel('x')
plt.legend()

plt.tight_layout()
plt.show()
```

Analyse des résultats
1. Impact du bruit
Bruit faible : Lorsque le bruit est faible, les données sont proches de la relation 
y=3x+2, et le modèle ajusté suit très bien cette relation. Le modèle a un biais faible et une faible variance.

Bruit moyen : Le bruit moyen commence à déformer le modèle ajusté. Le modèle commence à s'écarter de la vraie relation 
y=3x+2, ce qui montre un biais un peu plus élevé. La variance reste encore raisonnablement faible, mais le modèle est plus sensible aux petites fluctuations des données.

Bruit élevé : Lorsque le bruit est élevé, le modèle ajusté devient beaucoup plus instable, ce qui signifie une variance élevée. Le modèle commence à suivre trop de variations des données, ce qui peut entraîner un biais plus élevé et des prédictions moins fiables.

2. Biais et Variance

Biais : Le biais fait référence à l'écart entre la prédiction moyenne du modèle et la vraie valeur. Lorsque le bruit est faible, le biais est faible car le modèle peut bien ajuster les données. À mesure que le bruit augmente, le biais peut augmenter, car le modèle commence à faire des hypothèses incorrectes pour ajuster les fluctuations aléatoires dans les données.

Variance : La variance est la variation du modèle par rapport aux différentes données d'entraînement. Un modèle avec une haute variance ajustera trop les points de données (overfitting) lorsque le bruit est élevé, ce qui rendra le modèle sensible aux fluctuations spécifiques aux données d'entraînement.

3. Impact de la taille des données

La taille des données affecte la capacité du modèle à généraliser. Un petit échantillon peut entraîner un overfitting (sur-ajustement) ou un underfitting (sous-ajustement) selon la quantité de bruit dans les données. Un grand nombre de points permet généralement au modèle de mieux s'adapter aux tendances sous-jacentes, réduisant ainsi la variance et améliorant la stabilité des prédictions.

Conclusion

Un faible niveau de bruit donne un modèle avec un faible biais et une faible variance.
Un bruit plus élevé augmente la variance du modèle, ce qui peut conduire à un sur-ajustement des données.
La taille de l'échantillon affecte la capacité du modèle à généraliser et à réduire le biais et la variance.

```python
# Prédiction pour x1 = 28 ans, x2 = 4 heures
x_new = np.array([1, 28, 4])  # Inclut le biais
y_pred = x_new @ beta
print(f"Prédiction pour x1=28 et x2=4 : {y_pred:.2f} kg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Étape 1 : Données
x = np.array([1, 2, 3, 4, 5])  # Temps en jours
y = np.array([1.5, 3.0, 6.1, 12.2, 24.5])  # Population de la colonie bactérienne

# Transformation des données (logarithme de y)
y_transformed = np.log(y)

# Étape 2 : Ajustement de la régression linéaire
x_reshape = x.reshape(-1, 1)  # Transformation de x en vecteur colonne
model = LinearRegression()
model.fit(x_reshape, y_transformed)

# Coefficients du modèle
beta_0 = model.intercept_
beta_1 = model.coef_[0]

print(f"Intercept (β₀) : {beta_0:.2f}")
print(f"Coefficient (β₁) : {beta_1:.2f}")

# Étape 3 : Revenir à la forme exponentielle du modèle
# La relation exponentielle est : y = e^(β₀) * e^(β₁ * x)
def exponential_model(x, beta_0, beta_1):
    return np.exp(beta_0) * np.exp(beta_1 * x)

# Étape 4 : Prédiction pour x = 6
x_new = 6
y_pred = exponential_model(x_new, beta_0, beta_1)
print(f"Prédiction pour x = 6 jours : y = {y_pred:.2f} millions")

# Visualisation des données et de la droite de régression
plt.scatter(x, y, color='blue', label='Données observées')
x_line = np.linspace(1, 6, 100)
y_line = exponential_model(x_line, beta_0, beta_1)
plt.plot(x_line, y_line, color='red', label='Modèle de régression exponentielle')

plt.xlabel('Temps (jours)')
plt.ylabel('Population (millions)')
plt.legend()
plt.title('Régression exponentielle des données de population bactérienne')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Données avec point aberrant
x = np.array([25, 30, 35, 40, 45]).reshape(-1, 1)
y = np.array([2500, 3000, 3500, 4000, 8000])

# Régression avec point aberrant
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
r2_with_outlier = r2_score(y, y_pred)

# Données sans point aberrant
x_no_outlier = np.array([25, 30, 35, 40]).reshape(-1, 1)
y_no_outlier = np.array([2500, 3000, 3500, 4000])

# Régression sans point aberrant
model_no_outlier = LinearRegression()
model_no_outlier.fit(x_no_outlier, y_no_outlier)
y_pred_no_outlier = model_no_outlier.predict(x_no_outlier)
r2_without_outlier = r2_score(y_no_outlier, y_pred_no_outlier)

# Affichage des résultats
plt.figure(figsize=(12, 6))

# Avec point aberrant
plt.subplot(1, 2, 1)
plt.scatter(x, y, color="blue", label="Données")
plt.plot(x, y_pred, color="red", label="Régression")
plt.title(f"Avec point aberrant\nR² = {r2_with_outlier:.2f}")
plt.xlabel("Âge")
plt.ylabel("Salaire")
plt.legend()

# Sans point aberrant
plt.subplot(1, 2, 2)
plt.scatter(x_no_outlier, y_no_outlier, color="blue", label="Données")
plt.plot(x_no_outlier, y_pred_no_outlier, color="red", label="Régression")
plt.title(f"Sans point aberrant\nR² = {r2_without_outlier:.2f}")
plt.xlabel("Âge")
plt.ylabel("Salaire")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Données initiales
X = np.array([[25], [30], [35], [40], [45]])  # Âge
y = np.array([2500, 3000, 3500, 4000, 8000])  # Salaire

# Étape 1 : Ajuster une droite de régression linéaire
model = LinearRegression()
model.fit(X, y)

# Coefficients
beta_0 = model.intercept_  # Intercept (β₀)
beta_1 = model.coef_[0]    # Coefficient pour x (β₁)

# Prédictions
y_pred = model.predict(X)

# Calcul de R²
r2 = r2_score(y, y_pred)

print(f"Modèle avec le point aberrant:")
print(f"Intercept (β₀): {beta_0:.2f}")
print(f"Coefficient pour x (β₁): {beta_1:.2f}")
print(f"R²: {r2:.4f}")

# Étape 2 : Identifier le point aberrant
# Calcul des résidus
residuals = y - y_pred
print(f"Résidus : {residuals}")

# Affichage du modèle avec point aberrant
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Données")
plt.plot(X, y_pred, color="red", label="Régression (avec point aberrant)")
plt.title("Régression linéaire avec point aberrant")
plt.xlabel("Âge (ans)")
plt.ylabel("Salaire ($)")
plt.legend()
plt.show()

# Étape 3 : Réajuster le modèle sans le point aberrant
X_filtered = np.array([[25], [30], [35], [40]])  # Retirer x = 45
y_filtered = np.array([2500, 3000, 3500, 4000])

model_filtered = LinearRegression()
model_filtered.fit(X_filtered, y_filtered)

# Coefficients sans le point aberrant
beta_0_filtered = model_filtered.intercept_  # Intercept (β₀)
beta_1_filtered = model_filtered.coef_[0]    # Coefficient pour x (β₁)

# Prédictions sans le point aberrant
y_pred_filtered = model_filtered.predict(X_filtered)

# Calcul de R² sans le point aberrant
r2_filtered = r2_score(y_filtered, y_pred_filtered)

print("\nModèle sans le point aberrant:")
print(f"Intercept (β₀): {beta_0_filtered:.2f}")
print(f"Coefficient pour x (β₁): {beta_1_filtered:.2f}")
print(f"R²: {r2_filtered:.4f}")

# Affichage du modèle sans le point aberrant
plt.figure(figsize=(10, 6))
plt.scatter(X_filtered, y_filtered, color="blue", label="Données filtrées")
plt.plot(X_filtered, y_pred_filtered, color="green", label="Régression (sans point aberrant)")
plt.title("Régression linéaire sans point aberrant")
plt.xlabel("Âge (ans)")
plt.ylabel("Salaire ($)")
plt.legend()
plt.show()

# Comparaison des deux modèles
print("\nComparaison:")
print(f"Différence des coefficients β₁ : {abs(beta_1 - beta_1_filtered):.2f}")
print(f"Différence de R² : {abs(r2 - r2_filtered):.4f}")
```
