<span style="  
The function $\text({torch.corrcoef()})$ calculates the Pearson correlation matrix for the columns of a 2D tensor. In this context, the Pearson correlation between two variables $X $ and $ Y $ is given by the formula: 

$$
   r_{XY} = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^{n} (X_i - \bar{X})^2} \sqrt{\sum_{i=1}^{n} (Y_i - \bar{Y})^2}}

<span style=" 
where $\bar{X} $ and $ \bar{Y} $ are the means of $ X $ and $Y $, and $ n $ is the number of elements in each series.

<span style="

   $$

<La matrice de corrélation représente les relations linéaires entre toutes les colonnes du tenseur donné. Chaque élément $\rho_{ij} $ de la matrice est le coefficient de corrélation entre les colonnes  i  et j , et est calculé à l'aide de la formule suivante :

<s style=" 
The correlation matrix represents the linear relationships between all the columns of the given tensor. Each element $\rho_{ij}$ of the matrix is the correlation coefficient between columns $i$ and $j$, and is calculated using the following formula:

$$
\rho_{ij} = \frac{\text{cov}(X_i, X_j)}{\sigma_{X_i} \cdot \sigma_{X_j}}
$$

où :

-  $\text{cov}(X_i, X_j) $  est la covariance entre les colonnes $ X_{i} $ et $ X_{j} $,
- $ \sigma_{X_i} $ et $ \sigma_{X_j} $ sont les écarts-types des colonnes $ X_{i} $ et $ X_{j} $.

<L'écart-type d'une colonne $ X_{i} $ est défini comme la racine carrée de la variance de cette colonne :

<s style=" 
The standard deviation of a column $X_{i} $ is defined as the square root of the variance of that column:

$$
\sigma_{X_i} = \sqrt{\frac{1}{n} \sum_{k=1}^{n} \left( X_{i,k} - \bar{X}_i \right)^2}
$$

où :

-  $X_{i,k}$  est la $k^{-ième}$ valeur de $X_{i} $,
- $ \bar{X}_i $ est la moyenne des éléments de $ X_{i} $,
-  n  est le nombre d'éléments: n is the number of elements

$$
\text{cov}(X_i, X_j) = \frac{1}{n} \sum_{k=1}^{n} \left( X_{i,k} - \bar{X}_i \right) \left( X_{j,k} - \bar{X}_j \right)
$$

où : where

- $ X_{i,k} $ est la  $k^{-ième}$ valeur de la colonne $X_{i} $,
- $ \bar{X}_i $ est la moyenne des éléments de la colonne $ X_{i}$, calculée par :
    $$
    \bar{X}_i = \frac{1}{n} \sum_{k=1}^{n} X_{i,k}
    $$
- n  est le nombre d'éléments (lignes) dans chaque colonne:  <n  is the number of elements (rows) in each column.



<La matrice résultante est symétrique, avec des diagonales égales à 1, car la corrélation d'une colonne avec elle-même est toujours 1.

<s style=" 
The resulting matrix is symmetric, with diagonal elements equal to 1, since the correlation of a column with itself is always 1.

- Étapes pour obtenir la matrice de corrélation : <Steps to obtain the correlation matrix:


    - $\textbf{Standardisation des colonnes} $: <Standardization of the columns

    Pour chaque colonne $ X_i $, calculez la moyenne $\bar{X_i} $ et l'écart-type $\sigma_{X_i} $. Centrez et réduisez chaque colonne en soustrayant la moyenne et en divisant par l'écart-type.

    <s style=" For each column $ X_i $, calculate the mean $ \bar{X_i} $ and the standard deviation $ \sigma_{X_i} $. Center and scale each column by subtracting the mean and dividing by the standard deviation.

    - $\textbf{Calcul des produits croisés normalisés} $: <Calculation of normalized cross-products

    Pour chaque paire de colonnes $(X_i, X_j) $, calculez la covariance normalisée par les écarts-types. Cette opération donne les éléments  $\rho_{ij} $ de la matrice.

    <s style=" For each pair of columns $(X_i, X_j) $, calculate the covariance normalized by the standard deviations. This operation gives the elements $ \rho_{ij} $ of the matrix.

    - Construction de la matrice complète :  <Construction of the complete matrix.

    Répétez le calcul pour toutes les paires $ (i, j) $ où $ i, j \in \{1, 2, \dots, 6\} $, pour remplir la matrice $6 \times 6 $.
    
    <s style="   Repeat the calculation for all pairs $ (i, j) $ where $ i, j \in \{1, 2, \dots, 6\} $, to fill the 
    $ 6 \times 6 $ matrix.

- Matrice de corrélation résultante: Resulting correlation matrix.

Dans ce cas particulier, la matrice de corrélation a été calculée avec PyTorch. La matrice de corrélation $6 \times 6 $ résultante est :

<s style="

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Matrice de corrélation donnée: Given correlation matrix
correlation_matrix = np.array([
    [1.0000, -0.4015,  0.1834, -0.7634,  0.4550, -0.2301],
    [-0.4015,  1.0000, -0.1362,  0.1506,  0.3078, -0.4967],
    [0.1834, -0.1362,  1.0000,  0.2624, -0.2859,  0.1195],
    [-0.7634,  0.1506,  0.2624,  1.0000, -0.4262, -0.5197],
    [0.4550,  0.3078, -0.2859, -0.4262,  1.0000, -0.0085],
    [-0.2301, -0.4967,  0.1195, -0.5197, -0.0085,  1.0000]
])

# Labels des colonnes : Column labels
labels = [f"X{i+1}" for i in range(correlation_matrix.shape[0])]

# Création de la heatmap avec Seaborn : Creation of the heatmap with Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix, 
    annot=True,       # Affiche les valeurs sur la carte thermique: Display the values on the heatmap
    fmt=".4f",        # Format des nombres affichés:Number format displayed
    cmap="coolwarm",  # Palette de couleurs: Color palette
    xticklabels=labels,
    yticklabels=labels,
    vmin=-1, vmax=1   # Échelle des valeurs pour les corrélations:Scale of values for correlations
)

# Ajouter des titres : Add titles
plt.title("Matrice de Corrélation", fontsize=16) # Correlation Matrix
plt.xlabel("Colonnes", fontsize=12) # columns
plt.ylabel("Colonnes", fontsize=12)  # columns

# Afficher la carte thermique : Display the heatmap
plt.tight_layout()
plt.show()
```
