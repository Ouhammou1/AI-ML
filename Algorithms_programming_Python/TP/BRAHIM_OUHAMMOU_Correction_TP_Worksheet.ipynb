```python
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


image_path = '/content/exe1.png'
image = Image.open(image_path)
image = image.convert('RGB')
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
url = "https://raw.githubusercontent.com/chrisalbon/sim_data/master/titanic.csv"
titanic_df = pd.read_csv(url)

survivor_data = titanic_df.groupby(['PClass', 'Survived']).size().unstack()
survivor_data.plot(kind='bar', stacked=True, figsize=(10, 6), color=['red', 'green'])

plt.title("Number of Survivors and Non-Survivors by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.legend(['Non-Survivors', 'Survivors'], title="Survival Status")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
url = 'https://raw.githubusercontent.com/chrisalbon/sim_data/master/titanic.csv'
dataframe = pd.read_csv(url)

mean_age = dataframe['Age'].mean()
dataframe['Age'].fillna(mean_age, inplace=True)

plt.hist(dataframe['Age'], bins=20, color='blue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
url = "https://raw.githubusercontent.com/chrisalbon/sim_data/master/titanic.csv"
titanic_data = pd.read_csv(url)

titanic_data.rename(columns={"PClass": "PassengerClass"}, inplace=True)

age_stats = titanic_data.groupby("PassengerClass")["Age"].agg(
    Mean="mean",
    Median="median",
    Min="min",
    Max="max"
).reset_index()

print(age_stats)
```
